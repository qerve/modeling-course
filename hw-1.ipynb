### Решаем уравнение незатухающего маятника:

## $$ \ddot{\theta} + 2 \gamma \dot{\theta} + \omega_0^2 \sin{\theta} = 0$$


```python
z = 0.13
def f_acceleration(q, dot_q, gamma):
    ddot_q=-np.sin(q)-2*dot_q*gamma
    return ddot_q
```

### Функция f_acceleration принимает на вход положение, скорость и коэффициент затухания, а выводит ускорение $( \omega_0^2 = 1)$:
## $$ \ddot{\theta} = -  2 \gamma \dot{\theta} - \omega_0^2 \sin(\theta)  $$

```python
def f_acceleration(q, dot_q, gamma):
    ddot_q=-np.sin(q)-2*dot_q*gamma
    return ddot_q

fig, ax = plt.subplots()




def Euler(time, shagov, q_0, dot_q_0, z):
    a1 = [1]
    w1 = [0]
    q = q_0
    dot_q = dot_q_0
    h = time/shagov
    for i in range(shagov):
        ddot_q = f_acceleration(q, dot_q, z)
        b = q
        q = q + h*dot_q + h**2/2 * ddot_q
        dot_q = (q-b)/h
        a1.append(q)
        w1.append(dot_q)
        #print(q, dot_q, ddot_q)
        ax.plot(q, dot_q, '.', c='tab:blue')
        
    return q, dot_q, ddot_q
        
    
#f_acceleration(1, 0)
Euler(40, 40, 1, 0, 0.13)
ax.axhline(0, c='tab:red', label='Start point')
ax.axvline(1, c ='tab:red')
ax.plot(a1, w1, '-', label='Euler method')
ax.set_xlabel('$q, \; grad$')
ax.set_ylabel('$\dfrac{dq}{dt} , \; grad/s$')
ax.set_title('$\gamma = 0.13$')
ax.legend()
```

### а) Метод Эйлера

```python
t_span = [0, 40]
y0 = [1, 0]

def pendulum(t, y):
    g = 1
    l = 1
    theta, omega = y
    dydt = [omega, -g/l*np.sin(theta)]
    return dydt

sol_rk23 = solve_ivp(pendulum, t_span, y0, method='LSODA', min_step=1, max_step=1)
plt.plot(sol_rk23.y[0], sol_rk23.y[1], '-', label='RK23')
plt.legend()
```

```python
a2 = [1]
w2 = [0]
fig, ax = plt.subplots()

def nonEuler(time, shagov, q_0, dot_q_0, z):
    q = q_0
    dot_q = dot_q_0
    h = time/shagov
    for i in range(shagov):
        ddot_q = f_acceleration(q, dot_q, z)
        q1 = q + h*dot_q + h**2/2 * ddot_q
        dot_q1 = dot_q + h * ddot_q
        #ddot_q1 = f_acceleration(q1, dot_q1, 0.13)
        q = (q + q1) / 2
        dot_q = (dot_q + dot_q1) / 2
        a2.append(q)
        w2.append(dot_q)
        ax.plot(q, dot_q, '.', c='tab:blue')
    return q, dot_q, ddot_q
        
nonEuler(40, 40, 1, 0, 0.13)
ax.axhline(0, c='tab:red', label='Start point')
ax.axvline(1, c ='tab:red')
ax.plot(a2, w2, '-', label='Euler-Cromer method')
ax.set_xlabel('$q, \; grad$')
ax.set_ylabel('$\dfrac{dq}{dt} , \; grad/s$')
ax.set_title('$\gamma = 0.13$')
ax.legend()
```

```python
a3 = [1]
w3 = [0]
fig, ax = plt.subplots()

def Verlet(time, shagov, q_0, dot_q_0, z):
    q = q_0
    dot_q = dot_q_0
    h = time/shagov
    
    #  Euler first step
    ddot_q = f_acceleration(q, dot_q, z)
    q1 = 0.79 #q + h*dot_q + h**2/2 * ddot_q
    dot_q1 = -0.42 #dot_q + h * ddot_q
    #a3.append(q1)
    #w3.append(dot_q1)

    for i in range(shagov):
        ddot_q1 = f_acceleration(q1, dot_q1, 0.13)
        b = q1
        q2 = 2*q1 - q + ddot_q1 * h**2/2
        q = q1
        q1 = q2
        dot_q1 = (q1 - b)/h
        a3.append(q1)
        w3.append(dot_q1)
        ax.plot(q1, dot_q1, '.')
        
    return q, dot_q, ddot_q

Verlet(40, 40, 1, 0, z)
ax.plot(a3, w3, '-')
#ax.plot(a2, w2, '-')
#ax.plot(a1, w1, '-')
```

```python
def f_acceleration(q, dot_q, gamma):
    ddot_q=-np.sin(q)-2*dot_q*gamma
    return ddot_q

fig, ax = plt.subplots()

a5 = [1]
w5 = [0]

def Euler5(time, shagov, q_0, dot_q_0):
    q = q_0
    dot_q = dot_q_0
    h = time/shagov
    ddot_q = f_acceleration(q, dot_q, 0.13)
    q1 = q + h*dot_q + h**2/2 * ddot_q
    dot_q1 = dot_q + h * ddot_q
    for i in range(shagov):
        ddot_q1 = f_acceleration(q1, dot_q1, 0.13)
        q2 = q1 + h*dot_q1 + h**2/2 * ddot_q1
        dot_q1 = (q2-q)/(2*h)
        a5.append(q1)
        w5.append(dot_q1)
        q = q1
        q1 = q2
        #print(q, dot_q, ddot_q)
        ax.plot(q, dot_q, '.', c='tab:blue')
        
    return q, dot_q, ddot_q
        


#f_acceleration(1, 0)
Euler5(40, 40, 1, 0)
ax.axhline(0, c='tab:red', label='Start point')
ax.axvline(1, c ='tab:red')
ax.plot(a5, w5, '-', label='Euler method')
ax.set_xlabel('$q, \; grad$')
ax.set_ylabel('$\dfrac{dq}{dt} , \; grad/s$')
ax.set_title('$\gamma = 0.13$')
ax.plot(a1, w1, '-')
ax.legend()
```

```python
a6 = [1]
w6 = [0]
fig, ax = plt.subplots()

def Euler_Cramer(time, shagov, q_0, dot_q_0, z):
    q = q_0
    dot_q = dot_q_0
    h = time/shagov
    
    for i in range(shagov):
        ddot_q = f_acceleration(q, dot_q, z)
        #q1 = q + h*dot_q + h**2/2 * ddot_q
        dot_q1 = dot_q + h * ddot_q
        q1 = q + h*(dot_q1+dot_q)/2 + h**2/2 * ddot_q
        #ddot_q1 = f_acceleration(q1, dot_q1, 0.13)
        q = q1
        dot_q = dot_q1
        a6.append(q)
        w6.append(dot_q)
        ax.plot(q, dot_q, '.', c='tab:blue')
    return q, dot_q, ddot_q
        
Euler_Cramer(40, 40, 1, 0, z)
ax.axhline(0, c='tab:red', label='Start point')
ax.axvline(1, c ='tab:red')
ax.plot(a6, w6, '-', label='Euler-Cromer method')
#ax.plot(a1, w1, '-')
ax.set_xlabel('$q, \; grad$')
ax.set_ylabel('$\dfrac{dq}{dt} , \; grad/s$')
ax.set_title('$\gamma = 0.13$')
ax.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

### $$ x_{n+1} = 4r x_n (1 - x_n) $$

```python
def sequence(r, x_n, n, limit=1000):  # r, x_0, N
    if n == limit:
        return x_n
    else:
        x_next = 4 * r * x_n * (1 - x_n)
        return sequence(r, x_next, n+1, limit)

sequence(0.2, 0.4, 100)
```

```python
cols = ['red', 'blue', 'red']
r_tr = np.linspace(0, 1, 2000)
x_tr_1 = sequence(r_tr, 0.6, 500)
x_tr_2 = sequence(r_tr, 0.6, 501)
```

```python
fig, ax = plt.subplots()
ax.plot(r_tr, x_tr_1, '.', c='red')
ax.plot(r_tr, x_tr_2, '.', c='blue')
```

```python
plt.figure(figsize=(10, 7))
fig, ax = plt.subplots(figsize=(9, 7))

for i in range(10):
    ax.plot(r_tr, sequence(r_tr, 0.4, 500+i), '-', c=cols[(-1)**i])
    
#ax.plot(r_tr, sequence(r_tr, 0.4, int(502+(-1)**i/2)), '-', c=cols[(-1)**1])
```
